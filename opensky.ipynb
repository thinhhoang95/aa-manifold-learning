```python
from cartes.crs import Lambert93
from traffic.data import airports

with plt.style.context("traffic"):
    fig, ax = plt.subplots(subplot_kw=dict(projection=Lambert93()))
    for flight in sample_history:
        flight.plot(ax)
```
